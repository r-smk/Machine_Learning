```python
import seaborn as sns

df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
# finding the correlation between the independent features
corr=df.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
#jointplot

sns.jointplot(x='sepal_width',y='petal_length',data=df,kind='reg')
```

```python
sns.jointplot(x='petal_width',y='sepal_length',data=df,kind='hex')
```

```python
## distplot

sns.histplot(x='sepal_width',data=df,kde=True,bins=25)
```

```python
sns.histplot(x='petal_width',hue='species',data=df,kde=True,bins=25)
```

```python
sns.pairplot(data=df,hue='species')
```

```python
sns.boxplot(x='species',y='sepal_length',data=df)
```

```python
sns.boxplot(y='species',x='petal_width',data=df)
```

```python
sns.countplot(x='sepal_length',data=df)
```

```python
sns.countplot(y='petal_length',data=df)
```

```python
sns.barplot(y='sepal_length',x='petal_length',data=df)
```

```python
sns.violinplot(x='sepal_width',y='species',data=df)
```
