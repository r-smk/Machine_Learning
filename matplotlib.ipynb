```python
#Simple examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python

##Ploting using matplotlib

##plt scatter

plt.scatter(x,y,c='green')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')

plt.show()
```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,'ro',linestyle='dashed',linewidth=2,markersize=5)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plotted Graph')

plt.show()
```

```python
## creating subplots

# subplot(no.of rows,no.of cols, index)

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'go')
plt.subplot(2,2,4)
plt.plot(x,y,'y+')
```

```python
x=np.arange(1,11)
y=4*x**2+3*x+5
plt.plot(x,y,'r',linestyle='solid')
```

```python
x=np.arange(0,4*np.pi,0.0001)
y=np.sin(x)

plt.title('Sine wave form')

plt.plot(x,y)
```

```python
## Bar plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2)

plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
a = np.random.randint(0,1000,52,)
# print(a)
plt.hist(a,bins=80)
plt.title('Histogram')
```

```python
## pie chart

labels=['python','c++','ruby','java']
sizes=[215,130,245,210]
colors=['gold','skyblue','yellowgreen','lightcoral']
explode=(0.1,0,0,0)

#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
