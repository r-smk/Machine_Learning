```python
 import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr()
# we get only three types 
# because corr() takes only those columns which are integer or floating point values
```

```python
sns.heatmap(df.corr())
```

```python
# Univariate analysis

sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.distplot(df['tip'],kde=False,bins=20)
```

```python
sns.distplot(df['tip'],bins=20)
```
